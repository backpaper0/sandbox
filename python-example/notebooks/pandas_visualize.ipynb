# pandasでビジュアライズする

```python
import pandas as pd
from matplotlib import rcParams

# グラフの日本語が豆腐化するので日本語フォントを指定する。
# なんかPlemolJP以外は指定してもフォントを見つけてくれなかった。なぜ？
rcParams["font.family"] = "PlemolJP"

# デフォルトだとグラフはPNG形式で、Data URLが.ipynbファイルへ埋め込まれるみたい。
# グラフをSVGで出力したい場合は次のコードを実行する。
# %config InlineBackend.figure_format = 'svg'
```

兵庫県のオープンデータを利用して、地域毎のフリーWi-Fiスポット数を棒グラフで可視化してみる。

次のウェブページにある「設置箇所リスト（CSV：8KB）」を利用。

- [公衆無線LAN（Hyogo Free Wi-Fi）の運用・普及推進](https://web.pref.hyogo.lg.jp/kk26/musenlan.html)


```python
df = pd.read_csv("https://web.pref.hyogo.lg.jp/kk26/documents/hyogo_free_wi_fi_list_zahyo.csv")
```

```python
df["地域"].value_counts().plot.bar()
```
